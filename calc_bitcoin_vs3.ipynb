```python
df['ordem'] = range(0,len(df['Open']))
df['vrCloseOpen'] = df.Close-df.Open
df['vrHighLow'] = df.High-df.Low
df['mean72'] = df.Close.rolling(72).mean()
df['mean200'] = df.Close.rolling(200).mean()
df['closeGreaterM72'] = (df.Close>df.mean72)
df['closeGreaterM200'] = (df.Close>df.mean200)
df['M72GreaterM200'] = (df.mean72>df.mean200)
df['vrCloseM72'] = df.Close-df.mean72
df['vrCloseM200'] = df.Close-df.mean200
df['vrM72M200'] = df.mean72-df.mean200
df['tend72'] = np.array(df.Close.rolling(72).corr(df.ordem))
df['tend200'] = np.array(df.Close.rolling(200).corr(df.ordem))
df['dolar_vrUltimoAbertura'] = df.vl_ultimo-df.vl_abertura
df['dolar_vrMaximoMinimo'] = df.vl_maximo-df.vl_minimo
for i in range(1,80):
        df['vrHighDm'+str(i)] = df.Close - df.High.shift(periods=i)
        df['vrCloseDm'+str(i)] = df.Close - df.Close.shift(periods=i)
        df['dolar_vlMaximo'+str(i)] = df.vl_ultimo - df.vl_maximo.shift(periods=i)
        df['dolar_vlUltimo'+str(i)] = df.vl_ultimo - df.vl_ultimo.shift(periods=i)      
```

```python
import pandas as pd
dfPD = pd.DataFrame()
dfPD['High'] = df['High']/df['Close'] - 1
dfPD['Low'] = df['Low']/df['Close'] - 1
dfPD['Open'] = df['Open']/df['Close'] - 1
dfPD['Close'] = df['Open']/df['Close'] - 1
dfPD['Volume'] = df['Volume']/df['Volume'].rolling(72).mean() - 1
dfPD['vrCloseOpen'] = df['vrCloseOpen']/df['Close']
dfPD['vrHighLow'] = df['vrHighLow']/df['Close']
dfPD['mean72'] = df['mean72']/df['Close'] - 1
dfPD['mean200'] = df['mean200']/df['Close'] - 1
dfPD['closeGreaterM72'] = df['closeGreaterM72']
dfPD['closeGreaterM200'] = df['closeGreaterM200']
dfPD['M72GreaterM200'] = df['M72GreaterM200']
dfPD['vrCloseM72'] = df['vrCloseM72']/df['Close']
dfPD['vrCloseM200'] = df['vrCloseM200']/df['Close']
dfPD['vrM72M200'] = df['vrM72M200']/df['Close']
dfPD['tend72'] = df['tend72']
dfPD['tend200'] = df['tend200']
dfPD['diasem'] = df.index.dayofweek
dfPD['dolar_vrUltimoAbertura'] = df['dolar_vrUltimoAbertura']/df.vl_maximo.max()
dfPD['dolar_vrMaximoMinimo'] = df['dolar_vrMaximoMinimo']/df.vl_maximo.max()
for i in range(1,80):
        dfPD['vrHighDm'+str(i)] = df['vrHighDm'+str(i)]/df['Close']
        dfPD['vrCloseDm'+str(i)] = df['vrCloseDm'+str(i)]/df['Close']
        dfPD['dolar_vlMaximo'+str(i)] =  df['dolar_vlMaximo'+str(i)]/df.vl_maximo.max()
        dfPD['dolar_vlUltimo'+str(i)] = df['dolar_vlUltimo'+str(i)]/df.vl_maximo.max()   
```

```python
dfPD=dfPD.fillna(dfPD.mean())
```

```python
dfPD.head(10)
```

```python
import numpy as np
targetH = np.array(range(0,len(dfPD.index)))
i = 0
for idfPD in dfPD.index:
    targetH[i] = 0
    count = 1
    while count <= param.DayAhead:
        i2 = i + count
        if i2 >= len(dfPD.index): break
        if df.iloc[i2]['High']/df.iloc[i]['Close'] >= 1+param.gain: targetH[i] = 1
        count += 1
    i += 1
dfPD['targetH'] = targetH
```
